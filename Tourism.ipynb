**1.3.2.1**
There are 38 streets with counts, but only 34 counter locations. I want to compare counter locations with the the footfall counts, to see where these unlisted sensors are

```python
# create series to store Counter Locations
series_loc_counterLocations = pd.Series(counterLocations_df['Eco-Visio Oupput'])
series_loc_footfall23 = pd.Series(footfall23_df.columns)

# Convert Series to DataFrames
location_counterLocations = pd.DataFrame({'Counter Locations': series_loc_counterLocations})
location_footfall23 = pd.DataFrame({'Footfall Locations': series_loc_footfall23})

# Merge the DataFrames
location_comparison = pd.merge(location_counterLocations, location_footfall23, how='outer', left_on='Counter Locations', right_on='Footfall Locations')

# Sort the merged DataFrame alphabetically
location_comparison.sort_values(by='Counter Locations', inplace=True)

# Print the merged DataFrame
location_comparison
```

Here we can see that dawson street, dawson street old, liffey street old, and o'connell street penneys temperary locations are unlisted. these locations were also among the counters listed for missing Data for both St.Patricks day, and most of the year.
